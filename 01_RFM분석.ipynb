# 대형 마트 고객 데이터

## 지금부터 대형 마트 기업에 막 입사한 신입 데이터 분석입니다. 첫 업무는 2년치 고객 데이터를 활용해 여러그룹으로 고객으로 분류하고 이에 따라 마케팅 전략을 검토해보자. 분걱 결과를 바탕으로 핵심 고객 그룹을 집중 공략하면서 마케팅 자원 안에서 인사이트를 도출하자
## 고객을 여러 그룹으로 나누는 방법으로 RFM분석을 실시하는데 이는 얼마나 최근에 상품을 구매했는지, 얼마나 자주 상품을 구매했는지, 그리고 얼마나 돈을 많이 썼는지에 따라 고객을 분류하는 방법을 사용

## RFM 분석이란?

- Recency: 고객이 얼마나 최근에 상품을 구매했는가?
- Frequency: 고객이 얼마나 자주 상품을 구매했는가?
- Monetary: 고객이 상품 구매에 얼마나 많은 돈을 썼는가?

### RFM 분석을 한다는 건 위 세항목을 토대로 고객의 가치를 판단해 고객을 분류한다는 뜻입니다. 더 최근에 구매할수록, 더 자주 구매할수록, 돈을 더 많이 쓸수록 가치가 큰 고객으로 판단합니다. RFM 분석은 아주 기본적으로 효과적으로 고객을 ㅂ누류할 수 있는 방법입니다. 이렇게 분류된 그룹 하나하나를 세그먼트라고 합니다.

```python
import pandas as pd

ex1 = pd.DataFrame({
    'Recency' : ['3일', '100일', '40일', '5일'],
    'Frequency' : ['10회', '1회', '3회', '7회'],
    'Monetary' : ['100만원', '5만원', '150만원', '15만원']
})

ex1
```

###그리고 각각의 항목에 등급을 부여해야 합니다. 등급을 매기는 데에는 크게 3가지 방법이 있습니다.
###각 등급마다 고객의 수가 비슷해지도록 하거나, 구간의 길이를 똑같이 나누거나, 아니면 임의로 구간을 설정할 수도 있습니다.
###임의로 구간을 설정하는 예시를 들어볼게요. 각 항목의 등급을 3개로 나누되 아래처럼 기준을 정해 등급을 나누겠습니다.


- Recency : 7일 전까지는 3등급, 8일 전부터는 30일 전까지는 2긍급, 그 보다 과거는 1등급
- Frequency : 10회 이상은 3등급, 4회 이상 10회 미만은 2등급, 3회 이하는 1등급
- Monetary : 100 만원 이상은 3등급, 20만원 이상 100만원 미만은 2등급, 20만원 미만은 1등급

```python
ex1 = pd.DataFrame({
    'Recency' : ['3일', '100일', '40일', '5일'],
    'Recency등급' : ['3', '1', '1', '3'],
    'Frequency' : ['10회', '1회', '3회', '7회'],
    'Frequency등급' : ['3', '1', '1', '2'],
    'Monetary' : ['100만원', '5만원', '150만원', '15만원'],
    'Monetary등급' : ['3', '1', '3', '1']
})

ex1
```

###각 항목마다 3개의 등급이 있으니 총 27개의 세그먼트가 나옵니다
###이제 세그먼트별로 다른 전략을 세워 고객을 관리할 수 있습니다. 모든 등급이 높은 고객은 절대 놓치면 안 되니 수시로 쿠폰을 발급하거나 이벤트를 진행해 꾸준히 높은 매출을 올리게 해야 합니다. 반면 등급이 모두 낮은 고객은 이미 이탈했지만 딱히 아쉬울 것도 없는 고객이죠. Monetary 등급만 높고 Recency, Frequency 등급이 낮은 고객은 조금 더 자주 구매를 하도록 특별 쿠폰과 같은 방안을 모색해 보면 좋을 거예요. 반대로 Recency, Frequency 등급은 높은데 Monetary 등급이 낮은 고객은 충성도가 검증된 셈이니 상품을 맞춤 추천하여 추가 구매를 유도해 볼 수 있겠습니다.

###등급의 조합으로 고객 세그먼트를 나누면 세그먼트 개수는 이론적으로 등급 개수의 세제곱이 됩니다. 예시에서는 등급을 3개로 두어 27개의 세그먼트가 나왔는데요. 만약 등급을 5개로 둔다면 125개의 세그먼트가 나올 겁니다. 세그먼트 수가 많으면 고객 그룹을 더 세부적으로 파악하고 타깃을 정밀하게 설정할 수 있지만, 그만큼 세그먼트마다 고객을 관리하고 맞춤형 전략을 짜는 건 더 복잡해집니다. 일부 세그먼트는 너무 작은 규모여서 별로 의미가 없을 수도 있고, 비슷한 특성을 가진 고객이 서로 다른 세그먼트에 속하게 될지도 모르죠.

###이러한 한계를 보완하고자 각각의 등급으로 가중합을 구해 고객지표를 계산한 뒤, 이를 기반으로 다시 등급을 나누어 더 큰 단위의 고객 세그먼트를 구하기도 합니다. 먼저 고객 지표를 구하는 방법부터 보면 아래 식처럼 항목마다 가중치와 등급 값을 곱하여 더해 줍니다.

#$W_R$ * R + $W_F$ * F + $W_M$ * M

###가중치는 Recency, Frequency, Monetary 각 항목의 중요도를 고려하여 설정할 수 있습니다. 가령 배달 앱처럼 주문 금액의 상한이 그리 높지 않고 자주 이용하는 게 중요한 서비스에서는 Monetary에 비해 Recency나 Frequency가 더 중요한 항목이겠죠. 그러면 Monetary 가중치를 낮추고 다른 두 가중치를 높여서 고객 지표를 구하면 됩니다. 한편 가전제품은 주문당 결제 금액이 크고 한번 구매를 하면 재구매까지 시간이 오래 걸릴 테니, 판매사 입장에서는 다른 항목보다 Monetary 가중치를 상대적으로 높게 설정하는 게 좋을 거예요.

###이렇게 계산한 고객 지표를 적절한 구간으로 나누어 최종 RFM 등급을 구할 수 있습니다. RFM 등급이 곧 고객 세그먼트가 되겠죠. 고객 세그먼트를 몇 개로 나눌지는 직접 결정하면 됩니다. 이를테면 고객 지표의 범위가 0점에서 5점 사이로 나온다고 가정할게요. 세그먼트를 4개로 나누기로 하고, 4점 이상이면 VIP 등급, 3점 이상 4점 미만이면 골드 등급, 2점 이상 3점 미만이면 실버 등급, 그 밑으로는 브론즈 등급으로 분류해 볼 수 있겠죠. 고객 세그먼트가 수십 개씩 나올 때에 비해 복잡성이 줄고 맞춤형 전략을 수립하기도 쉬워질 겁니다.

###그럼 예시 데이터에서 고객 지표를 구하고 RFM 등급까지 부여해 세그먼트를 나눠 보겠습니다. 우선 Recency, Frequency, Monetary 세 항목이 똑같이 중요하다 가정할게요. 세 항목의 중요도가 전부 동일하다면 가중치 조건에 따라 세 가중치는 모두 1/3이 되어야 합니다. 이 가중치 세트로 고객 지표를 계산해 봅시다. 예를 들어 고객 A의 고객 지표는 1/3 * 3 + 1/3 * 3 + 1/3 * 3 = 3이 되는 거죠. 동일한 방식으로 나머지 고객들도 계산하면 다음과 같은 결과가 나옵니다.

```python
ex1 = pd.DataFrame({

    'Recency등급' : ['3', '1', '1', '3'],
    'Frequency등급' : ['3', '1', '1', '2'],
    'Monetary등급' : ['3', '1', '3', '1'],
    'RFM 고객 지표' : ['3', '1', '5/3', '2']
})

ex1
```

###이제 고객 지표에서 RFM 등급을 구할 차례입니다. RFM 등급 개수는 3개로 하고, 고객 지표가 1 이상 2 미만이면 1등급, 2 이상 3 미만이면 2등급, 3이면 3등급이라고 구간을 정해 볼게요. 그러면 RFM 등급이 아래처럼 매겨지게 됩니다. 예시 데이터에는 고객이 4명뿐이지만 실제로 수천 명 또는 수만 명의 고객이 단 3개의 세그먼트로 분류된다면 고객 관리가 훨씬 수월해지겠죠?

```python
ex1 = pd.DataFrame({
    'RFM 고객 지표' : ['3', '1', '5/3', '2'],
    'RFM 등급' : ['3', '1', '1', '2']
})

ex1
```

##매출 기여도를 고려한 가중치 조정
###앞서 가중치를 설정할 때 Recency, Frequency, Monetary 중에 더 비중을 두고 싶은 항목이 무엇인지에 따라 가중치를 설정하라고 했는데요. 이렇게 가중치 세트를 적용했는데 고객 세그먼트가 뚜렷하게 분류되지 않았다면 가중치를 조정하고 세그먼트를 재구성하는 게 좋습니다.

###여기서 뚜렷함의 기준이 되는 건 일반적으로 매출 기여도입니다. 매출 기여도는 말 그대로 매출에 얼마나 기여하는지를 나타내는 수치예요. 한 세그먼트의 매출 기여도는 전체 매출 대비 해당 세그먼트 고객의 매출 비율로 구할 수 있죠. 세그먼트마다 매출 기여도를 구해 보고 매출 기여도가 뚜렷하게 차이가 나는지, 아니면 모든 세그먼트의 매출 기여도가 비슷비슷한지 확인하면 됩니다.

###만약 VIP, 골드, 실버, 브론즈로 고객 등급을 분류했다면 VIP 등급의 매출 기여도가 확연히 높고 아래 등급으로 내려갈수록 매출 기여도가 낮아져야 좋겠죠. 그만큼 VIP 고객들이 특별한 가치가 있다는 뜻이고 이들을 집중적으로 관리하면 될 테니까요. 한편 각 등급의 매출 기여도가 비슷한 수준이라면 제대로 분류가 되지 않았다는 뜻이며 어느 세그먼트에 고객 관리 역량을 집중해야 할지 결정하기가 어려울 겁니다.

###가전제품 판매 회사에서 1년마다 RFM 분석을 통해 고객을 VIP, 골드, 실버, 브론즈 네 등급으로 분류하여 관리한다고 가정해 봅시다. 지난 1년의 데이터에 Recency, Frequency, Monetary 각 가중치를 1/3씩 균등하게 적용하여 세그먼트를 나눴더니 아래와 같은 결과가 나왔다고 해요. VIP 고객의 매출 기여도가 가장 높긴 하지만 브론즈 고객과의 차이가 6% 포인트에 불과합니다. 세그먼트마다 매출 기여도 차이가 별로 크지 않죠.

```python
ex1 = pd.DataFrame({
    '고객 등급' : ['VIP', '골드', '실버', '브론즈', '합계'],
    '매출' : ['2.8억', '2.5억', '2.5억', '2.2억', '10억'],
    '매출 기여도'  :['28%', '25%', '25%', '22%', '100%']
})

ex1
```

###가전제품의 특성을 고려해 Monetary 가중치를 0.6으로 높이고 Recency, Frequency 가중치는 각각 0.2로 낮춰 다시 세그먼트를 나눴더니 아래처럼 결과가 나왔다고 합니다. VIP 고객의 매출 기여도가 눈에 띄게 높아졌고 아래 등급으로 내려갈수록 매출 기여도가 크게 낮아지고 있죠. 이렇게 되면 VIP 고객이 명백히 핵심 고객인 셈이니 이들을 우선적으로 관리해 주면 될 거예요.

```python
ex1 = pd.DataFrame({
    '고객 등급' : ['VIP', '골드', '실버', '브론즈', '합계'],
    '매출' : ['6.4억', '2.2억', '1.1억', '0.3억', '10억'],
    '매출 기여도'  :['64%', '22%', '11%', '3%', '100%']
})

ex1
```
